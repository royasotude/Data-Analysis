```python
df = data.pivot_table(
    index='Year', 
    columns='Gender', 
    values='Medal', 
    aggfunc='count', 
    fill_value=0
)

df.columns = ["Men" , "Women"]

print(df)
```

```python
fig_gen = plt.figure(figsize = (10, 6))
ax_gen = plt.gca()
fig_gen.set(facecolor= "#EAF4f4")
ax_gen.set_facecolor("#EAF4f4")


ax_gen.plot(df['Men'], color = '#005f73', marker = 'o', label ='Men')
ax_gen.plot(df['Women'], color = '#94d2bd', marker = 'o' ,label = 'Women')

ax_gen.yaxis.tick_right()

ax_gen.fill_between(df.index , df['Men'], df['Women'], color = '#005f73', alpha = 0.6)
ax_gen.fill_between(df.index, df['Women'], color = '#94d2bd', alpha = 0.6)

ax_gen.tick_params(right = False, bottom = False , color= "#22333b")

ax_gen.spines['top'].set_visible(False)
ax_gen.spines['left'].set_visible(False)
ax_gen.spines['right'].set_visible(False)

plt.tick_params(axis = "y" , length =0 , labelcolor="#22333b")
plt.tick_params(axis = "x" , length =0 , labelcolor="#22333b")
plt.xlim(1896 , 2012)

plt.title('Male and Female Won Medals over the Years' , color= "#22333b" , fontsize=12)
plt.ylabel('Number of Won Medals' , color= "#22333b" , fontsize=10)
plt.legend(facecolor="#EAF4f4")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Olympics-Data.csv")
data.head()
```

```python
outer = data[data["Country"] == "SVK"].groupby("Sport").size()
outer
```

```python
inner = (data[data["Country"] == "SVK"]
        .groupby(["Sport", "Medal"])
        .size().unstack(fill_value=0))
inner
```

```python
outer_labels = outer.index.tolist()
outer_sizes = outer.values.tolist()
print(outer_labels , outer_sizes)
```

```python
inner_sizes = []

for sport in inner.index:
    for medal in inner.columns:
        value = inner.loc[sport , medal]
        if value > 0:
            inner_sizes.append(value)

inner_sizes , len(inner_sizes)
```

```python
outer_colors = ['#d8f3dc', '#b7e4c7', '#95d5b2','#74c69d','#52b788','#40916c']
```

```python
medal_colors = {'Gold': '#E9C46A','Silver': '#E0E1DD','Bronze': '#CD7F32'}
```

```python
inner_labels = []

for sport in inner.index:
    for medal in inner.columns:
        if inner.loc[sport , medal] > 0:
            inner_labels.append(medal)

inner_labels , len(inner_labels)
```

```python
inner_colors = []

for medal in inner_labels:
        inner_colors.append(medal_colors[medal])
inner_colors  , len(inner_colors)
```

```python
fig_SVK, ax_SVK =plt.subplots(figsize=(10,8))
fig_SVK.set(facecolor="#EAF4f4")

ax_SVK.pie(outer_sizes,
        radius=1.0,
        labels=outer_labels,
        colors=outer_colors,
        wedgeprops=dict(width=0.3, edgecolor='white'))


ax_SVK.pie(inner_sizes,
        radius=0.7,
        labels= inner_labels,
        colors=inner_colors,
        labeldistance=0.75,
        wedgeprops=dict(width=0.3, edgecolor='white'))

plt.title("Count of Medals in each Sport" , color = "#22333b" , fontsize=12)

plt.show() 
```
